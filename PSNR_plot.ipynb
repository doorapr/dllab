```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
psnr_values = pd.read_csv("average_psnr_overview.csv", index_col=0)
```

```python
psnr_values = psnr_values.rename(columns = {"masked_reconstructed_wheel_compared_to_original_wheel": "Wheel", "masked_reconstructed_bulldozer_compared_to_reconstructed_wheel" : "Reconstructed bulldozer to reconstructed wheel", "masked_reconstructed_bulldozer_compared_to_original_wheel": "Reconstructed bulldozer to original wheel", "masked_reconstructed_bulldozer_compared_to_original_bulldozer": "Bulldozer" })
psnr_values = psnr_values.rename(index = {"hd1": 1, "hd2" : 2, "hd4": 4, "hd8": 8, "hd16": 16, "hd32":32, "hd64": 64, "hd128": 128})

psnr_values = psnr_values.drop(columns=["Reconstructed bulldozer to reconstructed wheel"])
psnr_values.plot(style='.-')
plt.legend(loc=(0.3, 0.5))
plt.title('PSNR values for 2 datasets')
plt.xlabel('Hidden dimension')
plt.ylabel('PSNR')
plt.xscale("log", base=2)
plt.savefig("psnr_plot.pdf")
```
